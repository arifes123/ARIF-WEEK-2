It seems the label files are not organized in `train`, `val`, and `test` subdirectories within the `labels` folder. YOLO expects a similar structure for images and labels. Let's move them to the correct locations.

```python
!ls -lh /content/runs/detect/train4
```

```python
import os
import shutil

base_path = '/content/bubble_detection0'

# Create label subdirectories if they don't exist
for subset in ['train', 'val', 'test']:
    os.makedirs(os.path.join(base_path, 'labels', subset), exist_ok=True)

# Move label files to their respective subdirectories
for subset in ['train', 'val', 'test']:
    image_dir = os.path.join(base_path, 'images', subset)
    label_dest_dir = os.path.join(base_path, 'labels', subset)

    # Get all image files in the current subset directory
    image_files = [f for f in os.listdir(image_dir) if f.endswith(('.jpg', '.jpeg', '.png'))]

    for img_file in image_files:
        # Construct the corresponding label filename (e.g., image.jpg -> image.txt)
        label_file_name = os.path.splitext(img_file)[0] + '.txt'

        # Source path of the label file (all labels are currently in base_path/labels)
        label_src_path = os.path.join(base_path, 'labels', label_file_name)

        # Destination path for the label file
        label_dest_path = os.path.join(label_dest_dir, label_file_name)

        # Check if the label file exists in the source and move it
        if os.path.exists(label_src_path):
            shutil.move(label_src_path, label_dest_path)
            print(f"Moved {label_file_name} to {label_dest_dir}/")
        else:
            print(f"Warning: Label file {label_file_name} not found in {os.path.join(base_path, 'labels')}/")

print("Label files reorganized successfully!")
```

Now that the labels are correctly organized, you can re-run the YOLO training cell (cell with `!yolo train ...`). Make sure to use `device=cpu` since no CUDA devices are available, and `data=/content/bubble_detection0/bubbledetection.yaml` for the data configuration.
